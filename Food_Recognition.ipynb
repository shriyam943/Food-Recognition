```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

IMG_SIZE = 224
BATCH_SIZE = 32

train_datagen = ImageDataGenerator(

    rescale=1./255,

    rotation_range=20,

    zoom_range=0.2,

    horizontal_flip=True
)

test_datagen = ImageDataGenerator(
    rescale=1./255
)
```

```python
train_data = train_datagen.flow_from_directory(

    '/content/Fast Food Classification V2/Train',

    target_size=(224,224),

    batch_size=32,

    class_mode='categorical'
)
```

```python
test_data = test_datagen.flow_from_directory(

    '/content/Fast Food Classification V2/Test',

    target_size=(224,224),

    batch_size=32,

    class_mode='categorical'
)
```

```python
import tensorflow as tf
import numpy as np

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import GlobalAveragePooling2D

from tensorflow.keras.applications import MobileNetV2
```

```python
IMG_SIZE = 224
```

```python
base_model = MobileNetV2(

    weights='imagenet',

    include_top=False,

    input_shape=(224,224,3)
)
```

```python
base_model.trainable = False
```

```python
model = Sequential([

    base_model,

    GlobalAveragePooling2D(),

    Dense(128, activation='relu'),

    Dropout(0.3),

    Dense(train_data.num_classes, activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(

    optimizer='adam',

    loss='categorical_crossentropy',

    metrics=['accuracy']
)
```

```python
history = model.fit(

    train_data,

    validation_data=test_data,

    epochs=10
)
```

```python
model.save('/content/food_model.h5')
```

```python
loss, accuracy = model.evaluate(test_data)

print("Accuracy:", accuracy)
```

```python

import tensorflow as tf
import numpy as np
import pandas as pd
import cv2
from tensorflow.keras.models import load_model

# -----------------------------------
# Load Trained Model
# -----------------------------------
model = load_model('food_model.h5')

# -----------------------------------
# Nutrition Dataset
# -----------------------------------
nutrition_data = {

    "Food": [
        "Baked Potato",
        "Burger",
        "Crispy Chicken",
        "Donut",
        "Fries",
        "Hot Dog",
        "Pizza",
        "Sandwich",
        "Taco",
        "Taquito"
    ],

    "Calories": [
        161,
        354,
        320,
        452,
        365,
        290,
        285,
        250,
        226,
        312
    ],

    "Protein": [
        4,
        17,
        22,
        4,
        4,
        11,
        12,
        9,
        9,
        10
    ],

    "Carbs": [
        37,
        29,
        15,
        51,
        48,
        18,
        36,
        30,
        20,
        28
    ],

    "Fat": [
        0,
        19,
        18,
        25,
        17,
        19,
        10,
        8,
        13,
        18
    ]
}

nutrition_df = pd.DataFrame(nutrition_data)

# -----------------------------------
# Class Names
# -----------------------------------
class_names = [
    "Baked Potato",
    "Burger",
    "Crispy Chicken",
    "Donut",
    "Fries",
    "Hot Dog",
    "Pizza",
    "Sandwich",
    "Taco",
    "Taquito"
]

# -----------------------------------
# Streamlit Title
# -----------------------------------
st.title("AI Food Recognition System")

st.write("Upload a Food Image")

# -----------------------------------
# Upload Image
# -----------------------------------
uploaded_file = st.file_uploader(
    "Choose Food Image",
    type=["jpg", "jpeg", "png"]
)

# -----------------------------------
# Prediction
# -----------------------------------
if st.button("Predict Food"):

    if uploaded_file is not None:

        # Read Image
        file_bytes = np.asarray(
            bytearray(uploaded_file.read()),
            dtype=np.uint8
        )

        img = cv2.imdecode(file_bytes, 1)

        # Display Uploaded Image
        st.image(
            img,
            caption="Uploaded Image",
            use_container_width=True
        )

        # -----------------------------------
        # Preprocessing
        # -----------------------------------
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        img = cv2.resize(img, (224,224))

        img = img.astype('float32') / 255.0

        img = np.expand_dims(img, axis=0)

        # -----------------------------------
        # Prediction
        # -----------------------------------
        prediction = model.predict(img)

        predicted_class = np.argmax(prediction)

        food_name = class_names[predicted_class]

        # -----------------------------------
        # Nutrition Mapping
        # -----------------------------------
        food_info = nutrition_df[
            nutrition_df['Food'] == food_name
        ]

        calories = int(food_info['Calories'].values[0])

        protein = int(food_info['Protein'].values[0])

        carbs = int(food_info['Carbs'].values[0])

        fat = int(food_info['Fat'].values[0])

        # -----------------------------------
        # Recommendation Engine
        # -----------------------------------
        if calories > 400 or fat > 20:

            recommendation = "Eat Less"

            tip = "High calories detected. Reduce portion size."

        elif protein >= 15 and fat < 15:

            recommendation = "Healthy Choice"

            tip = "Good nutritional balance."

        else:

            recommendation = "Balanced Meal"

            tip = "Can be eaten in moderation."

        # -----------------------------------
        # Display Results
        # -----------------------------------
        st.success("Prediction Complete")

        st.subheader("Predicted Food")
        st.write(food_name)

        st.subheader("Nutrition Information")

        st.write("Calories:", calories, "kcal")

        st.write("Protein:", protein, "g")

        st.write("Carbs:", carbs, "g")

        st.write("Fat:", fat, "g")

        st.subheader("Health Recommendation")

        st.write(recommendation)

        st.info(tip)

    else:
        st.warning("Please upload an image")
```
